Analiz tarihimiz en son gün yapılan alışverişten 2 gün sonrası olacak şekilde yaparız.
master_id'ye göre gruplandırma yaparız. 
**Recency** --> Yenilik --> Analiz tarihinden müşterinin alışveriş yaptığı son tarih arasındaki fark bize recency değerini verir.
**Frequency** --> Sıklık --> Müşterinin ne kadar alışveriş yaptığı bize frequency'i değerini verir.

**Monetary** --> Parasal --> Toplam harcadığı tutarı verir.



```python
# GÖREV 2: RFM Metriklerinin Hesaplanması# GÖREV 2: RFM Metriklerinin Hesaplanması
#recency,frequnecy,monetary
df.head()
df["last_order_date"].max()  ## En büyük tarihi görürüz ve üzerine 2 gün ekleyerek analiz tarihini belirleriz.
today_date= dt.datetime(2021,6,1)
rfm=df.groupby("master_id").agg({"last_order_date": lambda date: (today_date - date.max()).days,
                                 "Omnichannel_Counts": lambda num: num,
                                 "Omnichannel_Price": lambda TotalPrice: TotalPrice.sum()})

rfm.columns=["Recency", "Frequency", "Monetary"]
rfm
```

**RFM-->** Müşteri segmentasyonu için kullanılan bir yöntemdir.
Müşterilerin satın alma alışkanlıkları üzerinden gruplara ayrılması ve bu gruplar özelinde stratejiler geliştirilmesini sağlar. Bu yüzden **recency ve frequency** değerleri bizim için önemlidir.
Recency değerinin küçük olması, frequency değerinin yüksek olması bizim için önemlidir.
**RFM-SKOR**--> Recency ve frequency değerleri bize rfm skorunu verir. Labels'lerde verdiğimiz değerler sonucunda rfm skorunu elde ederiz. 
Recency=5, Frequency=4 ise rsm skorumuz 54 şeklinde yazılır.




**pd.qcut-->** Değişkenleri küçükten büyüğe sıralar. q ve labelleri kendimiz veririz. 5 kırılım veriyorsak 5 label olmak zorundadır. q değeri kaç parçaya böleceğimizi, label ise alacağı değeri göstermektedir.
Değerler her kırılım içinde mevcut olabilir.Bu durumda rank methodunu kullanmak gerekir.


```python
# GÖREV 3: RF ve RFM Skorlarının Hesaplanması
rfm["recency_score"] = pd.qcut(rfm["Recency"], 5, labels=[5, 4, 3, 2,1])  
rfm["monetary_score"] = pd.qcut(rfm["Monetary"], 5, labels=[1, 2, 3, 4, 5])
rfm["frequency_score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5]) 
rfm["RFM_SCORE"] = (rfm["recency_score"].astype(str) + rfm["frequency_score"].astype(str))
rfm.head()
```

```python

# GÖREV 4: RF Skorlarının Segment Olarak Tanımlanması

seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'

}

rfm["segment"] = rfm["RFM_SCORE"].replace(seg_map, regex=True)
rfm=rfm[["Recency","Frequency","Monetary","segment"]]
```

```python
###GÖREV 5: Aksiyon zamanı!
           # 1. Segmentlerin recency, frequnecy ve monetary ortalamalarını inceleyiniz.
    
rfm.groupby("segment").agg({"Recency":"mean",
                            "Frequency":"mean",
                            "Monetary":"mean"})
```

Merge fonksiyonu ile datamızın ilk halinde "interested_in_categories_12" sütununu rfm isimli datamıza ekliyoruz. Bu eklemeyi iki data'da ortak olan master_id'ye göre yapıyoruz. how parametresi ile ne tarafa ekleyeceğimizi belirtiyoruz.
Sadık müşterileri(loyal_customers,champion) ve kadın kategorisinde alışveriş yapanları new_df'ye aktararak sonuçları csv olarak kaydediyoruz.


```python
# a. FLO bünyesine yeni bir kadın ayakkabı markası dahil ediyor. Dahil ettiği markanın ürün fiyatları genel müşteri tercihlerinin üstünde. Bu nedenle markanın
                   # tanıtımı ve ürün satışları için ilgilenecek profildeki müşterilerle özel olarak iletişime geçeilmek isteniliyor. Sadık müşterilerinden(champions,loyal_customers),
                   #  ve kadın kategorisinden alışveriş yapan kişiler özel olarak iletişim kuralacak müşteriler. Bu müşterilerin id numaralarını csv dosyasına
                   # yeni_marka_hedef_müşteri_id.cvs olarak kaydediniz.
            
merged_df = pd.merge(rfm, df[['master_id', 'interested_in_categories_12']], on='master_id', how='left')
new_df = merged_df[(merged_df["segment"].isin(["champions", "loyal_customers"])) &
                   (merged_df["interested_in_categories_12"].str.contains("KADIN"))]
new_df.to_csv("sonuclar.csv")
new_df
```
